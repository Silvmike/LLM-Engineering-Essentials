```python
import requests
import base64
from IPython.display import Image, display

# Replace with the actual ngrok URL printed in Step 1
server_url = "https://bbe1-34-48-111-201.ngrok-free.app/generate_and_describe"

# Make request
prompt = "A surreal city with floating islands and waterfalls"
response = requests.post(server_url, json={"prompt": prompt})
data = response.json()

# Show image and description
if 'image_base64' in data:
    image_data = base64.b64decode(data['image_base64'])
    display(Image(data=image_data))
    print("Description:", data['description'])
else:
    print("Error:", data.get('error'))
```
